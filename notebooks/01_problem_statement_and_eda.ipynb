### 5.4 Handle Outliers (if needed)

````python
# Check for outliers using IQR method
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    n_outliers = len(data[(data[column] < lower_bound) | (data[column] > upper_bound)])
    return n_outliers, lower_bound, upper_bound

# Get current numerical columns
current_numerical = df_clean.select_dtypes(include=[np.number]).columns.tolist()

print("Outlier analysis:")
print("="*60)
for col in current_numerical:
    n_outliers, lower, upper = find_outliers_iqr(df_clean, col)
    if n_outliers > 0:
        pct = n_outliers / len(df_clean) * 100
        print(f"{col}: {n_outliers:,} outliers ({pct:.1f}%) | bounds: [{lower:.2f}, {upper:.2f}]")
````

````python
# Handle outliers (if needed)
# TODO: Add your outlier handling code here if needed
#
# STRATEGIES:
# - Remove rows with outliers (be careful - losing data)
# - Cap/clip values at bounds
# - Keep them (if they're valid data points)
#
# Examples:
# df_clean = df_clean[df_clean['price'] > 0]  # Remove invalid prices
# df_clean = df_clean[df_clean['price'] < 500000]  # Remove extreme prices
# df_clean['column'] = df_clean['column'].clip(lower=0, upper=upper_bound)

# YOUR CODE HERE:

# Capping Hours_Studied
df_clean['Hours_Studied'] = df_clean['Hours_Studied'].clip(lower=4.0, upper=36.0)

# Capping Tutoring_Sessions (Lower bound set to 0 since negative sessions are impossible)
df_clean['Tutoring_Sessions'] = df_clean['Tutoring_Sessions'].clip(lower=0, upper=3.5)

# Capping Exam_Score
df_clean['Exam_Score'] = df_clean['Exam_Score'].clip(lower=59.0, upper=75.0)

print("Outliers handled in selected numerical columns.")
````

**Explain your outlier handling strategy (or why you kept them):**

I chose a Capping (Clipping) strategy to preserve the 6.5% of data that would have been lost if we simply deleted the outliers thats a large percentage of the data set. This method "squishes" extreme values to the statistical bounds, ensuring that rare scores don't disproportionately skew your averages or correlations. . It effectively cleans the "noise" while still keeping as much data as possible.
